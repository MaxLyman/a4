```python
df = pd.read_csv('games-data_new.csv')
df['r-date'] = pd.to_datetime(df['r-date'])
df['r-date'][0].year
df['user score'] = pd.to_numeric(df["user score"], errors='coerce')
df.dropna(inplace=True)
df.reset_index(inplace=True,drop=True)
df.head(50)
```

```python
sma_year = df.set_index('r-date').groupby('developer')['user score'].transform(lambda s: s.rolling(window='730D',closed = 'right').mean())

sma_year.reset_index(drop=True,inplace=True)
sma_year.reindex(df.index)
df['SMA_year'] = sma_year
df.loc[df['developer'] == 'LucasArts'].head(50)
df.tail(20)
```

```python
df.rename(columns={'r-date':'r_date'},inplace=True)

df = df[df.groupby('developer').developer.transform('count')>100].copy() 
df.to_csv('games_new_data.csv')
df.tail(20)
```

```python
platform = {}
all_platforms ={}
for row in df.itertuples():
    if row.platform not in all_platforms:
        all_platforms[row.platform] = 0
for row in df.itertuples():
    #if row[0] == 0:
     #   print(row._3.year)
        
    
    if row._3.year not in platform:
        if len(platform) == 0:
            
            platform_year = {}
            platform[row._3.year] = all_platforms.copy()
            platform[row._3.year][row.platform] = 0
        elif len(platform) != 0:
            platform[row._3.year] = platform[row._3.year-1].copy()
    if row._3.year in platform:
        platform[row._3.year][row.platform] +=1
        #if row.platform not in platform_year:
         #   platform_year[row.platform] = 1
        #elif row.platform in platform_year:
         #   platform_year[row.platform] +=1
list_of_df = []
for outer_dict in platform:
    platform_series = pd.DataFrame.from_dict(platform[outer_dict],orient='index', columns =['count'])
    platform_series['year'] = outer_dict
    platform_series.reset_index(level=0, inplace=True)
    
    list_of_df.append(platform_series)
platform_df = pd.concat(list_of_df)
platform_df.reset_index(drop=True,inplace=True)

platform_df['year'] = pd.to_datetime(platform_df['year'],format="%Y")
platform_df['year'].dt.strftime('%m/%d/%Y')
platform_df.rename(columns={'index':'name','count':'value','year':'date'},inplace=True)
platform_df.to_csv('platform.csv')
platform_df
```
