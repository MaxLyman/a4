```python
import pandas as pd
```

```python
df = pd.read_csv('games-data_new.csv')
df
```

```python
df['r-date'] = pd.to_datetime(df['r-date'])
df['r-date'][0].year
```

```python
platform = {}
all_platforms ={}
for row in df.itertuples():
    if row.platform not in all_platforms:
        all_platforms[row.platform] = 0
for row in df.itertuples():
    #if row[0] == 0:
     #   print(row._3.year)
        
    
    if row._3.year not in platform:
        if len(platform) == 0:
            
            platform_year = {}
            platform[row._3.year] = all_platforms.copy()
            platform[row._3.year][row.platform] = 0
        elif len(platform) != 0:
            platform[row._3.year] = platform[row._3.year-1].copy()
    if row._3.year in platform:
        platform[row._3.year][row.platform] +=1
        #if row.platform not in platform_year:
         #   platform_year[row.platform] = 1
        #elif row.platform in platform_year:
         #   platform_year[row.platform] +=1
list_of_df = []
for outer_dict in platform:
    platform_series = pd.DataFrame.from_dict(platform[outer_dict],orient='index', columns =['count'])
    platform_series['year'] = outer_dict
    platform_series.reset_index(level=0, inplace=True)
    
    list_of_df.append(platform_series)
platform_df = pd.concat(list_of_df)
platform_df.reset_index(drop=True,inplace=True)
platform_df.rename(columns={'index':'platform'},inplace=True)
platform_df.to_csv('platform.csv')
platform_df
```

```python
developer = {}
all_developer = {}
for row in df.itertuples():
    if row.developer not in all_developer:
        all_developer[row.developer] = 0
for row in df.itertuples():
    #if row[0] == 0:
     #   print(row._3.year)
        
    
    if row._3.year not in developer:
        if len(developer) == 0:
            
            developer_year = {}
            developer[row._3.year] = all_developer.copy()
            developer[row._3.year][row.developer] = 0
        elif len(developer) != 0:
            developer[row._3.year] = developer[row._3.year-1].copy()
    if row._3.year in developer:
        developer[row._3.year][row.developer] +=1
        #if row.developer not in developer_year:
         #   developer_year[row.developer] = 1
        #elif row.developer in developer_year:
         #   developer_year[row.developer] +=1
list_of_df = []
for outer_dict in developer:
    developer_series = pd.DataFrame.from_dict(developer[outer_dict],orient='index', columns =['count'])
    developer_series['year'] = outer_dict
    developer_series.reset_index(level=0, inplace=True)
    list_of_df.append(developer_series)
developer_df = pd.concat(list_of_df)
developer_df.reset_index(drop=True,inplace=True)
developer_df.rename(columns={'index':'developer'},inplace=True)

developer_df.to_csv('developer.csv')
developer_df
```
